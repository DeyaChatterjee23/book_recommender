```python
final_rating.drop_duplicates(['UID', 'Title'], inplace = True)
```

```python
final_rating.shape
```

```python
#Creating a pivot table
book_pivot = final_rating.pivot_table(columns='UID', index = 'Title', values = 'Ratings')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(book_pivot)
similarity_scores.shape
```

```python
def recommender(book):
    index=np.where(book.pivot.index==book)[0][0]
    distances=similarity_scores[index]
    similar_items=sorted(list(enumerate(distances)), key = lambda x:x[1], reverse=True)[1:6])
    suggestions=[]
    for i in similar_items:
        suggestions.append(book_pivot[i[0]])
    return suggestions
```

# Training the model

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
# Implementing Nearest Neighbours clustering algorithm to get the top recommendations
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors (algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
```

## Content based recommendation using LLM

```python
import pandas as pd
```

```python
filepath=r"/Users/deyachatterjee/Desktop/Book_recommender/Data/7kbooks.csv"
```

```python
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
def text_rep(row):
    tp=f"""Title: {row['title']}
    Authors: {row['authors']}
    Description: {row['description']}
    Categories: {row['categories']}
    Publishing year: {row['published_year']}
    Average Rating: {row['average_rating']}
    Number of pages: {row['num_pages']}
    """
    
    return tp
    
```

```python
df.iloc[[0]].apply(text_rep, axis=1).values[0]
```

```python
df['textual_representation']= df.apply(text_rep, axis=1)
```

```python
df.head()
```

```python
import faiss
import requests
import numpy as np
```

```python
dim=2304
index = faiss.IndexFlatL2(dim)
X = np.zeros((len(df['textual_representation']), dim), dtype = 'float32')
```

```python
for i, representation in enumerate(df['textual_representation']):
    if i % 100 == 0:
        print(i)
    res = requests.post('http://localhost:11434/api/embeddings',
                       json = {
                           'model': 'gemma2:2b',
                           'prompt': representation
                       })
    embedding = res.json()['embedding']
    X[i]= np.array(embedding)
index.add(X)
```

```python
faiss.write_index(index, 'index')
```

```python
index = faiss.read_index('index')
```
